```python
import pandas as pd
import os
import numpy as np
from shutil import copyfile
from tqdm import tqdm
import shutil
```

```python
df = pd.read_csv('finalStockPics.csv')
print(len(df))
df.head()
```

```python
tagsSet = set()
for tags in df['tags']:
    tags = tags.split(',')
    for i in range(len(tags)):
        tags[i] = tags[i].strip()
    tagsSet.update(tags)
try:
    tagsSet.remove('')
except:
    pass
print(len(tagsSet))
tagsSet
```

```python
if(not os.path.exists('dataset')):
    os.mkdir('dataset')
    for tag in tagsSet:
        try:
            os.mkdir('dataset/'+tag)
        except:
            pass
```

```python
err = 0
for data in tqdm(df.values):
    tags = data[0]
    tags = ['dataset/' + tag.strip() for tag in tags.split(',') if tag.strip()!='']
    imgSrc = data[1] 
    for i in tags:
        dst = i+"/"+imgSrc.split('/')[-1]
        try:
            copyfile(imgSrc, dst)
        except:
            err += 1
print(err)
# this code would have worked if I had the proper dataset and naming of images
```

```python
folders = os.listdir('dataset')
print(len(folders))
```

```python
folder_ = []
freq =[]
for folder in tqdm(folders):
    try:
        folder_.append(folder)
        freq.append(len(os.listdir('dataset/'+folder)))
    except:
        pass
```

```python
df = pd.DataFrame({'folder':folder_,'freq':freq})
df.sort_values(by='freq',ascending=False).head()
```

```python
df[df['freq']<50].head()
```

```python
try:
    os.mkdir('dataset/others')
except:
    pass
for i in tqdm(df[df['freq']<50]['folder'].values): # moving all folders with less than 50 images to a folder named 'others'
    src = 'dataset/'+i
    dst = 'dataset/others/'+i
    try:
        shutil.move(src,dst)
    except:
        pass
```
