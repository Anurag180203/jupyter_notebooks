```python
import requests
from tqdm import tqdm
import pandas as pd
from bs4 import BeautifulSoup
link = 'http://quotes.toscrape.com/'
page = requests.get(link)
page
```

```python
html = page.text
print(html)
```

```python
fd = open('main.html','w')
fd.write(str(html.encode('utf-8').strip()).replace('\\n','\n'))
fd.close()
```

```python
soup = BeautifulSoup(html,'html.parser')
print(soup.prettify())
```

```python
soup.find('span',class_='text').text[1:-1]
```

```python
list_of_quotes = []
for i in soup.find_all('span',class_='text'):
    list_of_quotes.append(i.text[1:-1])
list_of_quotes
```

```python
data = []
for sp in soup.find_all('div',class_='quote'):
    quote = sp.find('span',class_='text').text[1:-1]
    author = sp.find('small',class_='author').text
    author_id = sp.find('a').get('href')
    tags = sp.find('div',class_='tags').find_all('a',class_='tag')
    tags = ','.join([i.text for i in tags])
    print(quote,author,author_id,tags,sep='\n')
    print()
    data.append([quote,author,author_id,tags])
print(data.__len__())
```

```python
df = pd.DataFrame(data,columns=['quote','author','author_id','tags'])
df
```

```python
df.to_csv('quotes.csv',index=False)
```

```python
author_df = pd.read_csv('quotes.csv')
del author_df['quote']
del author_df['tags']
author_df = author_df.drop_duplicates()
author_df['author_id'] = link[:-1] + author_df['author_id'] 
author_df
```

```python
data = []
for author in tqdm(author_df.values):
    name = 'Authors/'+author[0]+'.txt'
    link = author[1]
    page = requests.get(link)
    soup = BeautifulSoup(page.text,'html.parser')
    text = soup.find('div',class_='author-details').text
    data.append(text.strip())
    fd = open(name,'w')
    fd.write(text)
    fd.close()
```

```python
author_df['author_info'] = data
author_df.head()
```

```python
author_df.to_csv('author_info.csv',index=False)
```
