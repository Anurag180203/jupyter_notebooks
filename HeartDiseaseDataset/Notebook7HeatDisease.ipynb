### Exploratory Data Analysis

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("heart.csv")
df.head()
```

```python
df.describe()
```

```python
sns.distplot(df["Age"])
```

```python
sns.distplot(df["Age"], kde=False, hist=True)
```

```python
sns.distplot(df["Age"], kde=False, hist=False,rug=True)
```

```python
sns.distplot(df["Age"], kde=True, hist=True,rug=True)
```

```python
sns.distplot(df["Age"],bins=30)
```

```python
sns.distplot(df["RestingBP"])
```

```python
sns.distplot(df["Cholesterol"])
```

```python
sns.distplot(df["MaxHR"])
```

```python
df.groupby('Sex').size().plot(kind='pie', autopct='%.2f%%')
```

```python
df.groupby('ChestPainType').size().plot(title="Chest Pain Types",kind='pie', autopct='%.2f%%')
```

```python
df.groupby('RestingECG').size().plot(title="Resting ECG",kind='pie', autopct='%.2f%%')
```

```python
df.groupby('ST_Slope').size().plot(kind='pie', autopct='%.2f%%')
```

```python
df.groupby('HeartDisease').size().plot(kind='pie', autopct='%.2f%%')
```

```python
df[df["Sex"]=='F'].groupby("HeartDisease").size().plot(kind="pie",autopct="%.2f%%")
```

```python
sns.violinplot(x="Age" , data=df)
```

```python
sns.violinplot(y="Sex", x="HeartDisease" , data=df)
```

```python
sns.violinplot(y="Age", x="HeartDisease" , data=df)
```

```python
sns.violinplot(y="RestingBP", x="HeartDisease" , data=df)
```

```python
sns.violinplot(y="Cholesterol", x="HeartDisease" , data=df)
```
