```python
import math
alpha = 0.05
def ztest(n, smaple_mean, pop_mean, pop_std,alternative):
    z_value = (smaple_mean - pop_mean)/(pop_std/math.sqrt(n))
    p_value = st.norm.sf(z_value)
    print(z_value, p_value,sep="\n")
    if alternative == "one-sided":
        if p_value < alpha:
            print("Reject Null Hypothesis")
        else:
            print("Fail to Reject Null Hypothesis")
    elif alternative == "two-sided":
        if p_value < alpha/2:
            print("Reject Null Hypothesis")
        else:
            print("Fail to Reject Null Hypothesis")
    

ztest(50, 110, 100, 15,"one-sided")
```

```python
sample_mean = 100
sample_sd= 20
data = sample_sd*np.random.randn(30)+sample_mean

alpha = 0.05
print("mean=%.2f stdv=%.2f" % (np.mean(data), np.std(data)))
t_statistic, p_value = st.ttest_1samp(a=data, popmean = 100)
print("t_statistic=%.2f, p_value=%.2f" % (t_statistic, p_value))
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

## Chi - Square Test

## ANOVA

```python
# checks if mean of 3 or more groups are equal or not while considering only one factor
english = [2,4,2]
maths = [2,3,4]
science = [1,2,5]
alpha = 0.05
stat,p_value = st.f_oneway(english,maths,science)
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
performance1 = [89, 89, 88, 78, 79]
performance2 = [93, 92, 94, 89, 88]
performance3 = [89, 88, 89, 93, 90]
performance4 = [81, 78, 81, 92, 82]
alpha = 0.05
stat,p_value = st.f_oneway(performance1,performance2,performance3,performance4)
if p_value < alpha:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
# for 2 factors 2 way ANOVA
import statsmodels.api as sm
from statsmodels.formula.api import ols
dataframe = pd.DataFrame({'Fertilizer': np.repeat(['daily', 'weekly'], 15),
                          'Watering': np.repeat(['daily', 'weekly'], 15),
                          'height': [14, 16, 15, 15, 16, 13, 12, 11, 14, 
                                     15, 16, 16, 17, 18, 14, 13, 14, 14, 
                                     14, 15, 16, 16, 17, 18, 14, 13, 14, 
                                     14, 14, 15]})
dataframe
```

```python
model = ols('height ~ Fertilizer + Watering + Fertilizer:Watering',data=dataframe).fit()
results = sm.stats.anova_lm(model, typ=2)
results
# as p values of both the factors are more than 0.05 we can say that both the factors are not significant and the same goes for their interaction
```
