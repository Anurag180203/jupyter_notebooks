```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,4*np.pi,50)
sin_y = np.sin(x)
cos_y = np.cos(x)
# print(x)
# print(sin_y)
plt.plot(x,sin_y,label="Sin(x)")
plt.plot(x,cos_y,label="Cos(x)")
plt.legend(loc="upper right")
plt.title("Sin and Cos")
plt.show()
```

```python
x = np.linspace(0,4*np.pi,50)
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,'--')
plt.legend(loc="upper right")
plt.title("Sin and Cos")
plt.show()
```

```python
x = np.linspace(0,4*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.figure(figsize=(15,5))
plt.style.use('Solarize_Light2')
plt.plot(x,y_sin,linewidth=3,color='blue',linestyle='-')
plt.plot(x,y_cos,color='red',linewidth="5",linestyle='--')
plt.title("Sine and Cosine")
plt.legend(['Sin(x)','Cos(x)'])
plt.show()
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.keys())
print(iris['target_names'])
x = iris.data
y=iris.target

plt.scatter(x[:,2],x[:,3],c="blue",s=50,alpha=.9)
plt.xlabel(iris["feature_names"][2])
plt.ylabel(iris["feature_names"][3])
plt.show()
```

```python
plt.scatter(x[:,2],x[:,3],c="blue",s=50,alpha=.9,marker='x')
plt.xlabel(iris["feature_names"][2])
plt.ylabel(iris["feature_names"][3])
plt.show()
```

```python
rmp = np.random.RandomState(0)
x = rmp.randn(100)
y = rmp.randn(100)
colors = rmp.rand(100)
sizes = rmp.rand(100)*1000

plt.scatter(x,y,c = colors,s=sizes,alpha=.7,cmap="viridis",edgecolors="red")
plt.title("Random scatter plot")
plt.xlabel('x values')
plt.ylabel('y values')
plt.colorbar()
plt.show()
```
