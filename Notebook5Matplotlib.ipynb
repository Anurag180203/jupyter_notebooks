```python
import matplotlib as mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,4*np.pi,50)
sin_y = np.sin(x)
cos_y = np.cos(x)
# print(x)
# print(sin_y)
plt.plot(x,sin_y,label="Sin(x)")
plt.plot(x,cos_y,label="Cos(x)")
plt.legend(loc="upper right")
plt.title("Sin and Cos")
plt.show()
```

```python
x = np.linspace(0,10,50)
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
fig,ax = plt.subplots(2,2)
ax[0,0].plot(x,sin_x)
ax[0,0].set_title("Sine wave")
ax[0,1].plot(x,cos_x)
ax[0,1].set_title("Cos wave")
ax[1,0].plot(x,tan_x)
ax[1,0].set_title("tan wave")
y=np.random.randn(1000)
ax[1,1].hist(y,bins=40)
ax[1,1].set_title("Random Distribution")
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0,10,50)
sin_x = np.sin(x)

for i in range(1,6):
    fig,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(18,6))
    ax1.plot(x,sin_x)
    ax2.plot(x,sin_x**2)
    ax3.plot(x,sin_x**3)
    ax1.set_title("Sine wave")
    ax2.set_title("Sine wave - Square")
    ax3.set_title("Sine wave - Cube")
plt.tight_layout()
plt.show()

# cos_x = np.cos(x)
# tan_x = np.tan(x)
# fig,ax = plt.subplots(2,2)
# ax[0,0].plot(x,sin_x)
# ax[0,0].set_title("Sine wave")
# ax[0,1].plot(x,cos_x)
# ax[0,1].set_title("Cos wave")
# ax[1,0].plot(x,tan_x)
# ax[1,0].set_title("tan wave")
# y=np.random.randn(1000)
# ax[1,1].hist(y,bins=40)
# ax[1,1].set_title("Random Distribution")
# plt.tight_layout()
# plt.show()
```
