```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Position_Salaries.csv')
data
```

```python
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
print(x,y,sep='\n')
```

### Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error

model = LinearRegression()
model.fit(x,y)
```

### Visualise the LR model

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, model.predict(x), color = 'blue')
plt.xlabel("input")
plt.ylabel("prediction")
plt.show()
```

### Polynomial Linear Regression

```python
poly_reg = PolynomialFeatures(degree=2)  # on increasing degree, accuracy increases
# but too much degree will lead to overfitting which is not good as it will decrease accuracy on test set
x_poly = poly_reg.fit_transform(x)
x_poly
```

```python
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly, y)
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_linear_reg.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.title("Polynomial Linear Regression")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=3) 
x_poly = poly_reg.fit_transform(x)
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly, y)
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_linear_reg.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.title("Polynomial Linear Regression degree=3")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=4) 
x_poly = poly_reg.fit_transform(x)
poly_linear_reg = LinearRegression()
poly_linear_reg.fit(x_poly, y)
plt.scatter(x, y, color = 'red')
plt.plot(x, poly_linear_reg.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.xlabel("Input")
plt.ylabel("Prediction")
plt.title("Polynomial Linear Regression degree=4")
plt.show()
```
